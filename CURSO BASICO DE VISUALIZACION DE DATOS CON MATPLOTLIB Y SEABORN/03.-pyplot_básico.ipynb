```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'bo') # grafica azul con bolas
plt.show()
```

```python
plt.plot(x,y, 'rx')  #grafica roja con x's
plt.show()
```

```python
plt.plot(x,y, 'rs-') #grafica roja con rombos y unidos por una linea
plt.show()
```

```python
plt.plot(x,y, 'yD:') #grafica amarilla con rombos y unidos por una puntos
plt.show()
```

```python
plt.hist(x) #grafica rx en un histograma
plt.show()
```

```python
plt.pie(x) #grafica de pie
plt.show()
```

```python
plt.scatter(x,y) #muestra relacion entre puntos o marquers entre nuetsras dos variabels
plt.show()
```

```python
plt.boxplot(x) 
plt.show()
```
