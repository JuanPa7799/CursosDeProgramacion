```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde',legend=True)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde',legend=True, palette='dark')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde',legend=True, palette='dark',alpha=.5)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex',legend=True, palette='dark',alpha=.2)
plt.show()
```
